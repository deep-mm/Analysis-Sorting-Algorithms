```python
import   pandas   as   pd
import   matplotlib.pyplot   as   plt
import   numpy   as   np
# Read the csv files
insertion   =   pd . read_csv ( 'insertion_block.csv' )
quick   =   pd . read_csv ( 'quick_block.csv' )
merge   =   pd . read_csv ( 'merge_block.csv' )
# Update the first column to be the block size from 1 to 10001 in steps of 1000
insertion [ 'n' ]   =   np . arange ( 1 ,   10001 ,   100 )
quick [ 'n' ]   =   np . arange ( 1 ,   10001 ,   100 )
merge [ 'n' ]   =   np . arange ( 1 ,   10001 ,   100 )

print ( insertion )
print ( quick )
print ( merge )
# Plot the graphs
# Make y axis runtime per comparison
plt . plot ( insertion [ 'n' ],   insertion [ 'runtime' ],   label = 'Insertion Sort' )
plt . plot ( quick [ 'n' ],   quick [ 'runtime' ],   label = 'Quick Sort' )
plt . plot ( merge [ 'n' ],   merge [ 'runtime' ],   label = 'Merge Sort' )
plt . xlabel ( 'block size' )
plt . ylabel ( 'runtime (ms)' )
plt . title ( 'Sorting Algorithms - Runtime - n (10000)' )
plt . legend ()
plt . show ()
```
